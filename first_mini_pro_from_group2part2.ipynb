```python
import pandas as pd

import ccxt
from datetime import datetime
import plotly.graph_objects as go
```

```python
binance = ccxt.binance()



trading_pair = 'BTC/USDT'
timeframe = '1d'
limit = 4000000
startDate = "2017-01-01"
startDate = datetime.strptime(startDate, "%Y-%m-%d")
startDate = datetime.timestamp(startDate)

startDate = int(startDate) * 1000
#print(startDate)

candles = binance.fetch_ohlcv(trading_pair, '1d' , since = startDate, limit=limit )
#candles[0]

dates = []
open_data = []
high_data = []
low_data = []
close_data = []

for candle in candles:
    dates.append(datetime.fromtimestamp(candle[0] / 1000.0).strftime('%Y-%m-%d %H:%M:%S.%f'))
    open_data.append(candle[1])
    high_data.append(candle[2])
    low_data.append(candle[3])
    close_data.append(candle[4])
```

```python
mongodata = []

for i in range(0,len(dates)):
  dictdata = {'_id' : dates[i] , 'open' : open_data[i]  , 'high' : high_data[i] , 'low' : low_data[i] ,  'close' : close_data[i] }  #

  mongodata.append(dictdata)
```

```python
fig = go.Figure(data=[go.Candlestick(x=dates,
                      open=open_data, high=high_data,
                      low=low_data, close=close_data)])
fig.show()
```
